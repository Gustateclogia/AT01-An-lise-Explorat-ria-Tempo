```python
tempo.loc[tempo['Aparencia'] == 'menos', 'Aparencia'] = 'sol'
```

```python
mediana_temp = tempo['Temperatura'].median() # Nota: ideal calcular a mediana sem o outlier
tempo.loc[tempo['Temperatura'] > 130, 'Temperatura'] = mediana_temp
```

```python
mediana_umid = tempo['Umidade'].median()
tempo['Umidade'].fillna(mediana_umid, inplace=True)
tempo.loc[(tempo['Umidade'] < 0) | (tempo['Umidade'] > 100), 'Umidade'] = mediana_umid
```
